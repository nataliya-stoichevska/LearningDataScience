# Data Tydying and Cleaning

# Tidy data
Правила за създаване и използване на дейтасети:

<strong>Колони</strong> - атрибути(особеност, променлива)

<strong>Редове</strong> - наблюдения

<strong>Клетки</strong> - стойности; хубаво е една клетка да има само една стойност (скаларна величина)

Всичко останало е messy data (неподредени данни)

Емпирично правило за тестване дали данните са подредени:

Ако добавим ново наблюдение трябва да добавим единствено един нов ред
 Не нови колони;
 Не повече редове;
 Не половинчати редове;
 Няма промени по съществуващи редове;
 
Важно е едно наблюдение да е уникално и да не се повтаря

```python
tb.sex_and_age.str.get(0).unique()
```

```python
tb["sex"] = tb.sex_and_age.str.get(0)
```

```python
tb
```

```python
tb["age_group"] = tb.sex_and_age.str.slice(1)
```

```python
tb
```

```python
tb.age_group.value_counts()
```

```python
tb.age_group.replace({"04": "0-4", "65": "65+", "u": "unknown"})
```

```python
def process_age_group(age_group):
    ages = {"04": "0-4", "65": "65+", "u": "unknown"}
    if age_group in ages:
        return ages[age_group]
    else:
        #Put a dash before the last two digits
        return f"{age_group[:-2]}-{age_group[-2:]}"
```

```python
tb.age_group = tb.age_group.apply(process_age_group)
```

```python
tb.age_group.unique()
```

```python
tb = tb.drop("sex_and_age", axis = 1) # axis=1 за изтриване по колони
#същото е tb.drop(columns = ["sex_and_age"])
```

```python
tb
```

```python
tb = tb.dropna() # маха липсващи стойностти na = not available хваща и  NAN
```

```python
tb.cases.unique() # уникални стойности
```

```python
tb.cases = tb.cases.astype(int)
```

```python
tb
```

```python
tb = tb[["iso2", "year", "sex", "age_group", "cases"]]
```

```python
tb = tb.sort_values(["iso2", "year"])
```

```python
tb.to_csv("data/tb_tidy.csv", index = None)
```

# Transformations on Features

The quality of our results depends strongly on the features we use
 "Garbage in – garbage out"
 
Dimensionality reduction

- Reducing the number of variables (features)

- We can do this manually or use algorithms

Feature selection - избираме колони, които ни трябват
- Selecting only columns that are useful

Feature extraction
- Transforming non-structured to structured data

Examples: images, audio, text

- Getting meaningful features

Feature engineering(най-важен процес) - комбинираме данни, колони, запълваме, създаваме нови...
- Using our knowledge of the data to create meaningful features
  - Involves a lot of brainstorming and testing


```python
weather["mean_temp_calculated"] = weather.min_temp + weather.max_temp / 2
```

```python
(weather.mean_temp - weather.mean_temp_calculated).min()
```
